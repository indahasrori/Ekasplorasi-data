# Eksplorasi Data Majalah Trend Motor US (Mtcars Dataset)

# Deskripsi

Data mtcars bersumber dari majalah Motor Trend US tahun 1974, yang berisi tentang data konsumsi bahan bakar dan 10 aspek desain serta kinerja untuk 32 mobil (model tahun 1973-1974).

# Variabel Penelitian

Data mtcars terdiri atas 11 variabel numerik dan 32 observasi pada setiap variabel.
- mpg (Miles/(US) gallon)
- cyl (Number of cylinders)
- disp (Displacement (cu.in.))
- hp (Gross horsepower)
- drat (Rear axle ratio)
- wt (Weight (1000 lbs))
- qsec (1/4 mile time)
- vs (Engine (0 = V-shaped, 1 = straight))
- am (Transmission (0 = automatic, 1 = manual))
- gear (Number of forward gears)
- carb (Number of carburetors)

# Import Modul/Package

```python
# Import Modul/Package
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import Data

```python
#Impor Data
data=pd.read_csv('mtcar.csv', sep=",")
data.head()
```

```python
data.info()
```

# Penambahan Variabel Baru (mpg_level)

```python
mpg=data['mpg'].values.tolist()
data['mpg_level']='a'
```

```python
for i in range (len(mpg)):
    if mpg[i] < 20:
        data['mpg_level'][i]='low'
    else:
        if mpg[i] < 30:
            data['mpg_level'][i]='medium'    
        else:
            data['mpg_level'][i]='hard'
```

```python
data.head()
```

```python
data.info()
```

# EKSPLORASI DATA

```python
#Mengurutkan Data Berdasarkan Variabel mpg
data_urut=data.sort_values(["mpg"]).reset_index(drop=True)
```

```python
data_urut.head()
```

```python
#Mean Variabel Mpg
Mean=data['mpg'].mean()
Mean
```

```python
# Barchart
fig, ax = plt.subplots(figsize=(15, 8))
ax.barh(data_urut['Cars'], data_urut['mpg'])
plt.axvline(Mean,color='red',label='Mean')
plt.title("Gas Miles for Car Models")
plt.xlabel("Miles/(US) gallon")
plt.ylabel("Cars")
```

```python
#Piechart
f,ax=plt.subplots(1,1,figsize=(8,5))
data['mpg_level'].value_counts().plot.pie(explode=[0,0.1,0.1],autopct='%1.1f%%',ax=ax,shadow=True)
ax.set_title('mpg_level')
ax.set_ylabel('')
plt.show()
```

Secara visual pada Barchart Gas Miles for Car Models, terdapat 14 mobil yang menggunakan bahan bakar diatas rata-rata atau seperti yang telah dijelaskan pada pie chart sebesar 56.2% mobil yang termasuk level low atau hemat bahan bakar, 31.2% mobil yang termasuk level medium, dan 12.5% mobil yang termasuk level hard atau boros bahan bakar. Selain itu, dapat diketahui bahwa mobil yang sedang trend di US tahun 1973-1974 dengan komposisi paling hemat bahan bakar adalah Toyota Corola, sedangkan mobil yang paling boros bahan bakar adalah Cadillac Fleetwood dan Lincoln Continental.

```python
#Matriks Korelasi
sns.heatmap(data.corr(),linewidths=.5,annot=True,cmap='RdYlGn')
fig=plt.gcf()
fig.set_size_inches(15,8)
plt.show()
```

Berdasarkan matriks korelasi mnunjukkan bahwa terdapat variabel yang mempunyai korelasi kuat dengan variabel mpg yaitu wt. Penggunaan bahan bakar (mpg) sangat dipengaruhi berat kendaraan (wt). Semakin berat kendaraan maka semakin kecil nilai mpg (Miles/(US) gallon) atau kendaran tersebut semakin boros bahan bakar dan sebaliknya. Akan tetapi, irit atau tidaknya penggunaan bahan bakar mobil tidak hanya ditentukan oleh berat kendaraan saja, melainkan gaya mengemudi seorang supir juga berperngaruh (qsec) meskipun nilai korelasinya tidak begitu kuat.

```python
#Boxplot
f,ax=plt.subplots(1,1,figsize=(18,8))
sns.boxplot(y="wt", x="cyl", hue="am", data=data, ax=ax)
fig=plt.gcf()
fig.set_size_inches(15,8)
ax.set_title('Boxplot Vihicle Weigh group by Miles/(US) gallon and Transmission')
plt.show()
```

Jenis transmisi menentukan berat kendaraan dimana mobil dengan  jenis transmisi automatic cenderung memiliki bobot yang lebih besar dibandingkan mobil dengan jenis transmisi manual. Berdasarkan Boxplot tersebut juga dapat diketahui bahwa mobil dengan jumlah silinder 6 memiliki variasi berat yang relatif homogen baik pada jenis transmisi automatic maupun manual artinya terdapat kesamaan kebutuhan bahan bakar antara jenis atau merk produk mobil yang satu dengan jenis mobil yang lainnya.

```python
#scatterplot
f,ax=plt.subplots(1,1,figsize=(18,8))
plt.scatter(x=data_urut['mpg'], y=data_urut['Cars'], c=data_urut['cyl'])
plt.title("Gas Miles for Car Models group by cylinder")
plt.xlabel("Miles/(US) gallon")
plt.ylabel("Cars")
plt.show()
```

Selain dipengaruhi oleh berat kendaraan, penggunaan bahan bakar juga ditentukan oleh jumlah silinder. Semakin banyak jumlah silinder pada mobil maka penggunaan bahan bakar akan semakin boros. Hal ini dikarenakan kapasitas mesin bisa mempengaruhi kinerja mobil dalam mengatur tenaga dan kecepatan. Jika kapasitas mesinnya besar, maka kendaraan tersebut mempunyai kecepatan tinggi dan membutuhkan tenaga yang besar.

```python
f,ax=plt.subplots(1,3,figsize=(18,8))
sns.boxplot(y="mpg", x="vs", data=data, ax=ax[0])
fig=plt.gcf()
fig.set_size_inches(16,5)
ax[0].set_title('Boxplot Miles/(US) by Engine')
sns.boxplot(y="mpg", x="carb", data=data, ax=ax[1])
fig=plt.gcf()
fig.set_size_inches(16,5)
ax[1].set_title('Boxplot Miles/(US) by Number of carburetors') 
sns.boxplot(y="mpg", x="am", data=data, ax=ax[2])
fig=plt.gcf()
fig.set_size_inches(16,5)
ax[2].set_title('Boxplot Miles/(US) by Transmission') 
plt.show()
```

Mobil dengan komponen penyusun mesin yang lurus (straight) lebih hemat dan banyaknya bahan bakar yang digunakan lebih beragam tergantung pada jenis mobil. Sealain itu, semakin sedikit jumlah karburetor mengindikasikan kendaraan tersebut akan hemat bahan bakar. Adapun bahan bakar yang digunakan oleh kendaraan dengan jumlah kerburetor 6 dan 8 lebih homogen artinya kendaraan jenis apapun yang terdiri dari 6 atau 8 komponen kerburetor mempunyai kebutuhan bahan bakar yang sama. Sedangkan untuk jenis mobil automatic lebih boros bahan bakar daripada mobil yang didesain dengan mesin manual.
