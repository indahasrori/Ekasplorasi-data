```python
# Import Modul/Package
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Impor Data
data=pd.read_csv('inflasi.csv', sep=",")
data.tail()
```

```python
data.describe()
```

```python
#Boxplot
dff=pd.melt(data[['Inflasi (%)']])
sns.boxplot(y="value", x="variable", data=dff)
fig=plt.gcf()
fig.set_size_inches(15,8)
plt.show()
```
